```
import urllib.request
import pandas as pd
import matplotlib.pyplot as plt

def main():
    gdpUrl = urllib.request.urlopen("https://raw.githubusercontent.com/datasets/gdp/0be54c18d900edc37123f25b4eff014731c9e459/data/gdp.csv")
    #read url to get csv
    gdpData = pd.read_csv(gdpUrl)#csv to DataFrame
    gdpData = gdpData.pivot(index="Year", columns="Country Code", values="Value")#restructure DataFrame
    #gdpData = gdpData[["CHN","EUU", "USA"]]#select only specified columns
    gdpCHN = gdpData["CHN"]
    gdpEUU = gdpData["EUU"]
    gdpUSA = gdpData["USA"]
    #separate data by column
    
    plt.plot(gdpCHN.index, gdpCHN.values/10**9, "r:")#plot China x=year, y=value/10^9, linestyle=red dot
    plt.plot(gdpEUU.index, gdpEUU.values/10**9, "g-.")#plot EU x=year, y=value/10^9, linestyle=green dashdot
    plt.plot(gdpUSA.index, gdpUSA.values/10**9, "b")#plot US x=year, y=value/10^9, linestyle=blue line
    plt.title("Michael Nguyen's GDP Plot")
    plt.legend(["CHN","EUU", "USA"])
    plt.xlabel("Year")
    plt.ylabel("GDP")
    plt.savefig('NguyenMichael_Skill6_Out.png')
    
if __name__ == '__main__':
    main()
```
